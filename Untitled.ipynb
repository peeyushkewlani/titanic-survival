```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the Titanic training dataset
train_data = pd.read_csv('train.csv')

# Load the Titanic testing dataset
test_data = pd.read_csv('test.csv')

# Load the gender submission file
gender_submission = pd.read_csv('gender_submission.csv')

# Select features and target variable
X_train = train_data[['Pclass', 'Sex', 'Age', 'Fare']]
y_train = train_data['Survived']

# Convert categorical variables to numerical using one-hot encoding
X_train = pd.get_dummies(X_train)
```

```python
# Initialize XGBoost classifier
model = XGBClassifier()

# Train the model
model.fit(X_train, y_train)

# Preprocess test data
X_test = test_data[['Pclass', 'Sex', 'Age', 'Fare']]
X_test = pd.get_dummies(X_test)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Create a DataFrame for submission
submission = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': y_pred})

# Save submission to a CSV file
submission.to_csv('submission.csv', index=False)
```

```python
# Plot bar chart of survivors
survived_counts = submission['Survived'].value_counts()
survived_counts.plot(kind='bar', color=['red', 'green'])
plt.style.use('dark_background')
plt.xlabel('Survival Status')
plt.ylabel('Passengers')
plt.title('#Survived')
plt.xticks([0, 1], labels=['0=Died', '1=Survived'], rotation=0)

#plt.legend(['0=Died', '1=Alive'])
#plt.legend(labels=labels_legend)
plt.legend(["0=Died", "1=Survived"], title_fontsize = 15,  prop = {'size' : 15})

plt.show()
```
